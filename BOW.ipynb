```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
doc=["people watch movie and movie", 
"people watch cricket and football",
"people watch football and movie",
"people watch movie and cricket",]
```

```python
bow=CountVectorizer()
bow_matrix=bow.fit_transform(doc)
```

```python
print(bow_matrix.toarray())
```

```python
# to get the feature names
print(bow.get_feature_names_out())
```
