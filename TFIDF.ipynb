```python
from sklearn.feature_extraction.text import TfidfVectorizer
doc=["people watch movie and movie",
"people like movie and cricket",
"people watch football and movie",
"people watch movie and cricket",
"people watch cricket and football"]
```

```python
tf_idf=TfidfVectorizer()
```

```python
vec=tf_idf.fit_transform(doc)
```

```python
print(vec.toarray())
```

```python
for vector in vec.toarray():
    print(vector)
```

```python
print("Feature names:", tf_idf.get_feature_names_out())
```
